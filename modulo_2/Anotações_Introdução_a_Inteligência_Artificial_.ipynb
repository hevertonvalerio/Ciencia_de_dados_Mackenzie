Antes de inicar a aula responder as seguintes questões:

AI (Artificial Intelligence)

*   O que é?
   * Inteligência Artificial é uma macro categoria que engloba áreas de estudo como Redes Neurais, Visão Computacional, Machine Learning. As áreas relacionadas à Inteligência Artificial são responsáveis pelo desenvolvimento de soluções através de algorítimos computacionais.
*   Qual é a sua definição de IA?
   * IA é o termo utilizado pra se referir à sistemas computacionais e matemáticos, de algoritimos complexos, organizados para a otimização de processos ou a solução de determinados problemas.  
*   Onde a IA pode ser Aplicada?
   * Não há limites para a utilização da inteligência artificial, assim como áreas básicas de estudo como matemática ou administração, a IA possui ilimitadas aplicações na sociedade uma vez que seu objetivo seja justamente unir artificios matemáticos e gerenciais para otimizar processos decisórios.
