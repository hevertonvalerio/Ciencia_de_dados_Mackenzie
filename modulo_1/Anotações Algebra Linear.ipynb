**Objetivo da aula**

Estudar o conceito de autovetores de uma transformação linear, vetores cuja imagem pela transformação consiste em um múltiplo dele próprio, e seus autovalores associados. Relembrar alguns detalhes sobre determinantes de matrizes e os usar para buscar autovalores e autovetores de uma transformação linear. Por fim, verificar como simplificar a representação de uma transformação linear que tenha uma base de autovetores.


```python
# prompt: importar biblioteca que calcula autovalores e autovetores

import numpy as np

# Calculate eigenvalues and eigenvectors of a square matrix
def eig(A):
  """
  Calculates the eigenvalues and eigenvectors of a square matrix.

  Args:
    A: A numpy array representing a square matrix.

  Returns:
    w: A numpy array containing the eigenvalues of A.
    V: A numpy array containing the eigenvectors of A as columns.
  """

  # Check if the matrix is square
  if A.shape[0] != A.shape[1]:
    raise ValueError("Matrix must be square.")

  # Calculate the characteristic polynomial
  p = np.poly(A)

  # Find the roots of the characteristic polynomial
  w = np.roots(p)

  # Find the eigenvectors for each eigenvalue
  V = np.zeros((A.shape[0], len(w)))
  for i in range(len(w)):
    V[:, i] = np.linalg.solve(A - w[i] * np.eye(A.shape[0]), np.ones(A.shape[0]))

  return w, V
```
