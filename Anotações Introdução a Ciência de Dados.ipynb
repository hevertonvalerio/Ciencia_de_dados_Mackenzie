domain é dominio sobre a área em questão

Em uma primeira análise, podemos entender que as habilidades necessárias
para formar um bom profissional da Ciência de Dados demandarão o aprendizado
em três grandes domínios:
- Engineering Skills (Computação).
- Estatística.
- Experiência de Domínio.
- Experiência de Negócios.


Além das relações que convergem para a Ciência de Dados no centro do dia-
grama, indicando que todas as habilidades destacadas são necessárias, temos
também as relações paralelas entre as principais habilidades.


Por exemplo, o diagrama destaca habilidades de Engenharia de Computação,
na qual buscaremos conhecimentos sobre algoritmos, lógicas, habilidades de pro-
gramação e domínio sobre o hardware no qual as aplicações rodam. Mas também
observamos que os algoritmos são parte do domínio da estatística e da mineração
de dados, claro porque ambas se valem dos algoritmos para implementar os mo-
delos que são utilizados com essas habilidades.


 mesmo ocorre quando observamos as relações entre Engenharia e expertise
ou experiência do domínio. No caso do domínio, estamos falando de um domínio
de conhecimento, uma área como Medicina, Química, Engenharia etc. Lembrando
aqui que as aplicações da Ciência de Dados podem ser implementadas em qualquer
domínio, mas, para isso, precisamos conhecer e analisar os sistemas envolvidos. Por esse motivo, temos os sistemas analíticos relacionados entre essas duas habilidades.


Por fim, a relação entre o domínio e a mineração de dados nos dá o destaque re-
lacionado à pesquisa, a qual também é uma possibilidade profissional. Neste caso,
temos o termo “ciência” usado mais adequadamente para o profissional cientista
que fará pesquisas sobre novos modelos de busca de padrões, classificadores ou
técnicas associadas a área.


Falando mais diretamente sobre as habilidades principais, pensando no âmbito
da Computação, agregaremos os conhecimentos sobre programação e lógica, que
serão um apoio muito importante para dominarmos a construção de sistemas ana-
líticos. Além disso, o conhecimento sobre hardware é importante para entender-
mos os limites computacionais, em termos de processamento e armazenamento
de dados.


Da Matemática e Estatística, agregaremos elementos importantes para o do-
mínio da análise exploratória de dados, bem como de modelos estatísticos que
podem ser muito úteis para o entendimento de fenômenos e para a otimização de
processos. Soma-se, aqui, a possibilidade de simular situações e compor cenários
que permitirão maior clareza sobre as decisões a serem tomadas por gestores.


O último item, nos revela a importância de conhecer o domínio de origem dos
dados. Muitos algoritmos que executaremos geram um resultado numérico. Saber
como isso se aplica ao que estamos estudando é fundamental. Por isso, temos que
incorporar o negócio ao estudo para entender melhor o que os dados representam.


 modo complementar, habilidades pessoais em comunicação, apresentação
e curiosidade também são muito importantes devido à complexidade envolvida e
às relações entre tecnologias, processos e pessoas. Esta relação é conhecida como abordagem sociotécnica.


A abordagem sociotécnica é uma conhecida área da formação em sistemas de
informação. Mas, a afinidade destas relações com o que fazemos na Ciência de
Dados nos faz refletir sobre o quanto devemos nos preocupar com as pessoas e os
processos, quando temos a tecnologia como ferramenta de atuação para nossos
estudos analíticos.


**Big Data**

O termo “Big Data” começa a aparecer em nossas vidas aproximadamente em
2010
O primeiro documento que discute o tema foi escrito pelo Gartner, que sugeria um fenômeno relacionado ao alto VOLUME de dados, VELOCIDADE e VARIEDADE da informação, demandando formas inovadoras de processar esses dados na tomada
de decisões. O Gartner chamou isso de 3Vs e, posteriormente, seria chamado de
Big Data.


O NIST (2015) define Big Data como a inabilidade das arquiteturas tradicionais
em trabalhar de modo eficiente com os novos conjuntos de dados, devido às carac-
terísticas de **VOLUME, VELOCIDADE, VARIEDADE, VERACIDADE**, conhecidas como
 **os Vs do Big Data** . Portanto, os conjuntos de dados são extensivos datasets que requerem
uma arquitetura computacional capaz de escalar tarefas de armazenamento, ma-
nipulação e análise.
